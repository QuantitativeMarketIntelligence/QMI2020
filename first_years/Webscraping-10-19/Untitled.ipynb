```python
import csv
import tqdm

# writing to csv file
with open('goodreads.csv', 'w') as csvfile:
    columns = ['Title', 'Author']
    csvwriter = csv.writer(csvfile)
    
    for i in range(1, 10):
        page = requests.get("https://www.goodreads.com/list/show/143444.Best_books_of_2020?page=" + str(i))
        soup = BeautifulSoup(page.content, 'html.parser')
        names = soup.find_all('span', itemprop='name')
        #write to csv
        idx = 0
        while idx < len(names) - 1:
            csvwriter.writerow([names[idx].text, names[idx + 1].text])
            idx += 2
```
